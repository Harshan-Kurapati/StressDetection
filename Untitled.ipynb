#  Correlations and Relationships

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x= "screen_on_time", y = "sleep_duration", data = averaged_df)
plt.title("Does screen on time impacts sleep duration?")
plt.show()

print(f"The correlation between Screen on time and Sleep Duration is : {averaged_df['screen_on_time'].corr(averaged_df['sleep_duration'])}")
```

# PSS_score vs. PSQI_score 

```python
plt.figure(figsize = (10,5))

plt.scatter(averaged_df["PSS_score"], averaged_df["PSQI_score"])
plt.title("PSS_score vs. PSQI_score ")

print(f"The correlation score between 'PSS_score' and 'PSQI_score': {averaged_df['PSS_score'].corr(averaged_df['PSQI_score'])}")
```

# Grouping by Personality

```python
for trait in ["Openness", "Conscientiousness", "Extraversion", "Agreeableness", "Neuroticism"]:
    median_value = averaged_df[trait].median()
    averaged_df[f'{trait}_group'] = averaged_df[trait].apply(lambda x: 'High' if x > median_value else 'Low')
    
```

```python
openness_group_analysis = averaged_df.groupby("Openness_group")["sleep_duration", "screen_on_time"].mean()

openness_group_analysis
```

```python
plt.figure(figsize = (10,5))
sns.boxplot(x = "Openness_group", y = "sleep_duration", data = averaged_df)
```

```python
import seaborn as sns

crosstab = pd.crosstab(averaged_df['Openness_group'], averaged_df['Agreeableness_group'])
sns.heatmap(crosstab, annot=True, cmap="YlGnBu", fmt='d')
plt.title('Heatmap of Openness and Agreeableness Groups')
plt.xlabel('Agreeableness Group')
plt.ylabel('Openness Group')
plt.show()
```
